```python
import pandas as pd
import numpy as np
```

```python
# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv('datos.csv')

# Convertir la columna 'Rating' a numérica, manejando los errores
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')


# Filtrar los registros que tienen un Rating válido (no nulos)
df_clean = df[df['Rating'].notna()]

# Verificar que se han filtrado correctamente los registros sin Rating
df_clean.head()
```

```python
# Agrupar por 'Course Title' y calcular el promedio de las calificaciones
promedio_calificaciones = df_clean.groupby('Course Title')['Rating'].mean()

# Mostrar el promedio de calificaciones
promedio_calificaciones
```

```python
# Encontrar el índice del curso con mayor calificación
curso_mayor_rating = df_clean.loc[df_clean['Rating'].idxmax(), 'Course Title']

# Encontrar el índice del curso con menor calificación
curso_menor_rating = df_clean.loc[df_clean['Rating'].idxmin(), 'Course Title']

# Mostrar los resultados
print(f"Curso con mayor rating: {curso_mayor_rating}")
print(f"Curso con menor rating: {curso_menor_rating}")
```

```python
# Filtrar los cursos con horario flexible
cursos_flexibles = df[df['Schedule'] == 'Flexible schedule']

# Calcular el porcentaje de cursos flexibles en relación al total
porcentaje_flexibles = (len(cursos_flexibles) / len(df)) * 100

# Redondear el porcentaje a dos decimales
porcentaje_flexibles = np.round(porcentaje_flexibles, 2)

# Mostrar el porcentaje
print(f"Porcentaje de cursos con horario flexible: {porcentaje_flexibles}%")
```
